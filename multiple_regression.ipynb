```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

computers = pd.read_csv("/Users/davidosullivan/Desktop/Regression Analysis/Course Materials/Data/Computers.csv")

computers.tail()
```

```python
sns.heatmap(computers.corr(numeric_only=True), annot=True, vmin=-1, vmax=1, cmap='coolwarm' )
```

```python
sns.pairplot(computers, corner=True)
```

```python
cols = ['ram', 'speed', 'hd', 'screen']

X = sm.add_constant(computers[cols])
y = computers['price']

model = sm.OLS(y,X).fit()

model.summary()
```

## Scatter Plot Interpretation

The below visual indicates that our model tends to be more accurate as price goes up. The points being highlighted indicate how far off the model was. The closer the points are to zero the better.

Wide range of error, in some circumstances we are off by more than 2k. Potentially some additional features to reduce the margin for error.

```python
sns.scatterplot(x=model.predict(), y=model.resid)
```

```python
cols = ['ram', 'speed', 'hd', 'screen', 'ads', 'trend']

X = sm.add_constant(computers[cols])
y = computers['price']

model = sm.OLS(y,X).fit()

model.summary()
```

## Continued Analysis 
Including more features in the model increases our r-squared and reduces the range of errors.

```python
sns.scatterplot(x=model.predict(), y=model.resid)
```
